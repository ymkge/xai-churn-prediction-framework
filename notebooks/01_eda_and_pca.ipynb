# 01. EDA and PCA
このノートブックでは、データの基本統計量の確認とPCAを用いた高次元データの可視化を行います。

```python
import sys
sys.path.append('../')
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from src.preprocessing import load_data, preprocess_data

df = load_data('../data/sample_churn_data.csv')
df_preprocessed = preprocess_data(df)
df_preprocessed.head()
```

```python
# 離脱(churn)別の分布可視化
sns.pairplot(df_preprocessed[['avg_login_per_month', 'learning_progress', 'churn']], hue='churn')
plt.show()
```

```python
# PCAによる2次元可視化
X = df_preprocessed.drop('churn', axis=1)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

plt.scatter(X_pca[:, 0], X_pca[:, 1], c=df_preprocessed['churn'], alpha=0.5)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA Projection')
plt.show()
```
