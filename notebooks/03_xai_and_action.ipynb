# 03. XAI and Action
SHAP, ICE, DiCEを用いて、予測の根拠と具体的な改善施策を検討します。

```python
import sys
sys.path.append('../')
import lightgbm as lgb
import dice_ml
from src.preprocessing import load_data, preprocess_data, split_data
from src.visualization import plot_shap_summary, plot_ice, show_dice_counterfactuals

df = load_data('../data/sample_churn_data.csv')
df_preprocessed = preprocess_data(df)
X_train, X_test, y_train, y_test = split_data(df_preprocessed)

model = lgb.LGBMClassifier(random_state=42)
model.fit(X_train, y_train)
```

## SHAP Summary Plot

```python
plot_shap_summary(model, X_test)
```

## ICE Plot

```python
plot_ice(model, X_test, ['avg_login_per_month', 'learning_progress'])
```
